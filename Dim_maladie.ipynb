```python
import os
import fitz

def extract_text_from_pdf(pdf_path):
    try:
        doc = fitz.open(pdf_path)
        text = ""
        for page in doc:
            text += page.get_text()

        text = text.replace('\n', ' ')  # Replace newline characters with spaces
        return text
    except Exception as e:
        print(f"Error while extracting text from {pdf_path}: {e}")
        return None

def extract_text_from_folder(folder_path):
    pdf_texts = []
    for filename in os.listdir(folder_path):
        if filename.endswith(".pdf"):
            pdf_path = os.path.join(folder_path, filename)
            extracted_text = extract_text_from_pdf(pdf_path)
            if extracted_text is not None:
                pdf_texts.append(extracted_text)
    return pdf_texts

folder_path = "/content/drive/MyDrive/pdf_files"
pdf_texts = extract_text_from_folder(folder_path)
print(pdf_texts)
```

```python
from nltk.tokenize import sent_tokenize

def print_sentences(pdf_texts):
    for index, extracted_text in enumerate(pdf_texts, start=1):
        sentences = sent_tokenize(extracted_text)
        print(f"Sentences in text{index}:")
        for sentence in sentences:
            print(sentence)
        print()  # Adding an empty line between texts for clarity

print_sentences(pdf_texts)
```

```python
import pandas as pd
from nltk.tokenize import sent_tokenize

def extract_questions_and_answers(pdf_texts):
    questions_and_answers = {'question': [], 'answer': []}

    for index, extracted_text in enumerate(pdf_texts, start=1):
        sentences = sent_tokenize(extracted_text)

        # Iterate through sentences
        i = 0
        while i < len(sentences):
            if sentences[i].endswith('?'):
                # Found a question
                question = sentences[i]
                answer = ""
                i += 1
                # Append sentences to answer until the next question is found
                while i < len(sentences) and not sentences[i].endswith('?'):
                    answer += sentences[i] + " "
                    i += 1
                questions_and_answers['question'].append(question.strip())
                questions_and_answers['answer'].append(answer.strip())
            else:
                i += 1

    return pd.DataFrame(questions_and_answers)

# Assuming pdf_texts is already defined
df = extract_questions_and_answers(pdf_texts)
print(df)
```
